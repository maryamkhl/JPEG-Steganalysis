```python
from scipy.io import loadmat
import scipy.io
import h5py
import os
import glob
import cv2
import numpy as np
import numpy
from skimage.util.shape import view_as_blocks, view_as_windows
from keras.utils import np_utils
import random
```

```python
def load_images1(path_pattern):

    files=glob.glob(path_pattern + '/*.mat')
    Y=[]
    for f in files:
        mat1 = scipy.io.loadmat(f)
        mat = np.array(mat1['matOH'])
        print(mat.shape)
        Y.append(mat)

    Y=numpy.array(Y)

    return Y
```

```python
path = 'bossbase_256_75_onehot_nsf5-0.5_9component_T10' 
Xc = load_images1(path + '/train/clean/')
Xs = load_images1(path + '/train/stego/')
Yc = load_images1(path + '/test/clean/')
Ys = load_images1(path + '/test/stego/')
Zc = load_images1(path + '/validation/clean/')
Zs = load_images1(path + '/validation/stego/')

X = numpy.vstack((Xc, Xs))
Y = numpy.vstack((Yc, Ys))
Z = numpy.vstack((Zc, Zs))


Xt = numpy.hstack(([0]*len(Xc), [1]*len(Xs)))
Yt = numpy.hstack(([0]*len(Yc), [1]*len(Ys)))
Zt = numpy.hstack(([0]*len(Zc), [1]*len(Zs)))


Xt = np_utils.to_categorical(Xt, 2)
Yt = np_utils.to_categorical(Yt, 2)
Zt = np_utils.to_categorical(Zt, 2)

idx=list(range(len(X)))
random.shuffle(idx)


X=X[idx]
Xt=Xt[idx]
```

```python
import keras
from keras.layers import Lambda, Layer
from keras.models import Sequential, Model,load_model
from keras.layers import Dense, Dropout, Activation, Flatten, LSTM
from keras.layers import Conv2D, MaxPooling2D, AveragePooling2D, GlobalAveragePooling2D, UpSampling2D
from keras.layers import Input, Convolution2D, LeakyReLU, Softmax
from keras.layers import concatenate
# from keras.layers.convolutional import AtrousConvolution2D
from keras.layers.core import Reshape
from keras import optimizers
from keras.layers.normalization import BatchNormalization
from keras import backend as K
from keras import initializers
from keras.optimizers import SGD
from keras import regularizers
from keras import initializers


# def build_model(num_classes):

#     input = Input(shape=(64,64,40))

#     # Conv Layer 1, Feature learning
#     layer_1_conv = Conv2D(
#         256,
#         kernel_size=1,
#         padding='same',
#         strides=2,
#         activation = 'relu'
#     )(input)
    
#     layer_2_conv = Conv2D(
#         256,
#         kernel_size=3,
#         activation = 'relu',
#         padding='same',
#         strides=2)(input)
    
#     layer_3_conv = Conv2D(
#         256,
#         kernel_size=5,
#         padding='same',
#         activation = 'relu',
#         strides=2)(input)

#     c = concatenate([layer_1_conv, layer_2_conv, layer_3_conv])
    
#     GA = GlobalAveragePooling2D()(c)
#     FC1 = Dense(units=400)(GA)
#     dropout1 = Dropout(0.2)(FC1)
#     FC2 = Dense(units=200)(dropout1)
#     dropout2 = Dropout(0.2)(FC2)
#     FC3 = Dense(units=2)(dropout2)
#     dropout3 = Dropout(0.2)(FC3)
#     output = Softmax()(dropout3)
    
#     return Model(inputs=input, outputs=output)



def build_model(num_classes):

    input = Input(shape=(32,32,90))

    # Conv Layer 1, Feature learning
    layer_1_conv = Conv2D(
        128,
        kernel_size=1,
        padding='same',
        strides=2,
        activation = 'relu',
        kernel_initializer=keras.initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None)
    )(input)
    
    layer_2_conv = Conv2D(
        128,
        kernel_size=3,
        padding='same',
        strides=2,
        activation = 'relu',
        kernel_initializer=keras.initializers.TruncatedNormal(mean=0.0, stddev=0.05, seed=None)
    )(input)

    c = concatenate([layer_1_conv, layer_2_conv])
    
    GA = GlobalAveragePooling2D()(c)
    dropout1 = Dropout(0.12)(GA)
    FC = Dense(units=2)(dropout1)
    dropout2 = Dropout(0.12)(FC)
    output = Softmax()(dropout2)
    
    return Model(inputs=input, outputs=output)
```

```python
num_classes=2
model=build_model(num_classes)
model.summary()
```

```python
from keras.optimizers import SGD , RMSprop , Adadelta, Adam, Adamax
sgd = SGD(lr=1e-3, decay=1e-6, momentum=0.9, nesterov=False) #better1
Optimizer=RMSprop(lr=0.0001) #better2
Adadelta = keras.optimizers.Adadelta(lr=0.05) #better3-epoch38
adam = Adam(lr=1e-3, epsilon = 1e-8, beta_1 = .9, beta_2 = .999) #better4
adamax = Adamax(lr=0.001, beta_1=0.9, beta_2=0.999, epsilon=1e-07)
model.compile(optimizer=sgd, loss='binary_crossentropy', metrics=['accuracy'])
```

```python
history = model.fit(X, Xt,
                    epochs=100,
                    batch_size=1,
                    validation_data=(Z, Zt))
```

```python
import matplotlib.pyplot as plt

acc = history.history['accuracy']
val_acc = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(len(acc))

plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()

plt.figure()

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.show()
```

```python
print("Evaluate on test data")
results = model.evaluate(Y, Yt,  batch_size=1)
print("test loss, test acc:", results)
```
